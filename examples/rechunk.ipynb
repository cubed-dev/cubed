# Rechunk with Cubed

The example here is based on the dataset size from [this Cubed issue](https://github.com/cubed-dev/cubed/issues/859).

```python
import cubed
import cubed.array_api as xp
```

```python
shape = (394488, 778, 706)
source_chunks = (24, 778, 706)
target_chunks = (43800, 5, 5)
```

```python
spec = cubed.Spec(allowed_mem="2GB")

a = xp.empty(shape, dtype=xp.float32, chunks=source_chunks, spec=spec)
a
```

```python
b = a.rechunk(target_chunks)
b
```

```python
b.visualize()
```

### Create random data

```python
import cubed.random
from cubed.diagnostics.rich import RichProgressBar
```

```python
a = cubed.random.random(shape, dtype=xp.float32, chunks=source_chunks)
a
```

```python
b = a.rechunk(target_chunks)
b
```

```python
b.visualize()
```

```python
with RichProgressBar():
    cubed.to_zarr(b, store=None)
```

### Workaround

```python
intermediate_chunks = (3650, 50, 50)
```

```python
a = cubed.random.random(shape, dtype=xp.float32, chunks=source_chunks)
a
```

```python
b = a.rechunk(intermediate_chunks)
b
```

```python
b.visualize()
```

```python
c = b.rechunk(target_chunks)
c
```

```python
c.visualize()
```
