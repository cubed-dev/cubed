# Pangeo Vorticity Workload

This is a notebook for exploring a simplified version of the example in https://github.com/pangeo-data/distributed-array-examples/issues/1.

In particular, it is the code from https://github.com/coiled/coiled-runtime/issues/174

```python
import cubed.array_api as xp
import cubed.random
from cubed.extensions.tqdm import TqdmProgressBar
```

```python
a = cubed.random.random((1000, 900, 800), chunks=100)
b = cubed.random.random((1000, 900, 800), chunks=100)
x = cubed.random.random((900, 800), chunks=100)
y = cubed.random.random((900, 800), chunks=100)

result = a[1:] * x + b[1:] * y
result = xp.mean(result)
```

```python
result.visualize("pangeo-vorticity")
```

```python
progress = TqdmProgressBar()
result.compute(callbacks=[progress])
```
