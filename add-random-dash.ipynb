```python
import cubed
import cubed.array_api as xp
import cubed.random
from cubed.diagnostics.dash import Dashboard
```

```python
a = cubed.random.random((25000, 25000), chunks=(5000, 5000))
b = cubed.random.random((25000, 25000), chunks=(5000, 5000))
c = xp.add(a, b)
```

```python
with Dashboard(debug=True):
    cubed.to_zarr(c, store=None, optimize_graph=False)
```
